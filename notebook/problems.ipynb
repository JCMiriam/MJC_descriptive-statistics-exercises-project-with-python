### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import statistics as stats
import pandas as pd
from scipy.stats import kurtosis, skew

def get_metrics(arr):
    return {
        'mean': np.mean(arr),
        'median': np.median(arr),
        'var': np.var(arr),
        'std': np.std(arr),
        'mode': stats.mode(arr),
        'min': np.min(arr),
        'max': np.max(arr),
        'range': np.max(arr) - np.min(arr),
        'skew': skew(arr),
        'kurt': kurtosis(arr),
    }

normal_distribution = np.random.normal(size=100)
chisquare_distribution = np.random.chisquare(df=3, size=100)

normal_metrics = get_metrics(normal_distribution)
chisquare_metrics = get_metrics(chisquare_distribution)

data = {
    'Metrics': ['Mean', 'Median', 'Variance', 'Standard Deviation', 'Mode', 'Min', 'Max', 'Range', 'Skew', 'Kurtosis'],
    'Normal Distribution': [
        normal_metrics['mean'],
        normal_metrics['median'],
        normal_metrics['var'],
        normal_metrics['std'],
        normal_metrics['mode'],
        normal_metrics['min'],
        normal_metrics['max'],
        normal_metrics['range'],
        normal_metrics['skew'],
        normal_metrics['kurt']
    ],
    'Chi-square Distribution': [
        chisquare_metrics['mean'],
        chisquare_metrics['median'],
        chisquare_metrics['var'],
        chisquare_metrics['std'],
        chisquare_metrics['mode'],
        chisquare_metrics['min'],
        chisquare_metrics['max'],
        chisquare_metrics['range'],
        chisquare_metrics['skew'],
        chisquare_metrics['kurt']
    ],
}

df = pd.DataFrame(data)
df
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO
import math

def get_std(data):
    mean = sum(data) / len(data)
    squared_diff = [(x - mean) ** 2 for x in data]
    sum_squared_diff = sum(squared_diff)
    variance = sum_squared_diff / len(data)
    std = math.sqrt(variance)

    return std

data = [4, 2, 5, 8, 6]

std = get_std(data)
std
````
